## Data cleaning

### Cleansing

The method to remove many unnecessary words from the document is to remove HTML characters, RT, URL, hashtag (#), mention username (@username), symbols, emoticons, two-letter spaces, and one-letter characters.

```python
import re

def cleansing(text):
    # Remove HTML tags
    text = re.sub(r'<.*?>', '', text)
    # Remove RT (Retweets)
    text = re.sub(r'\bRT\b', '', text)
    # Remove URLs
    text = re.sub(r'http\S+|www.\S+', '', text)
    # Remove hashtags
    text = re.sub(r'#\w+', '', text)
    # Remove mentions
    text = re.sub(r'@\w+', '', text)
    # Remove symbols and emoticons
    text = re.sub(r'[^\w\s]', '', text)
    # Remove one-letter characters (excluding 'I' and 'a' for readability)
    text = re.sub(r'\b(?!I\b|a\b)\w\b', '', text)
    # Replace multiple spaces with a single space
    text = re.sub(r'\s+', ' ', text).strip()
    return text

df['cleaned_text'] = df['text'].apply(cleansing)
df[['text','cleaned_text']].head()
```

### Case Folding 

This process is used to convert uppercase letters 
into lowercase letters.

```python
df['cleaned_text'] = df['cleaned_text'].apply(lambda x: x.lower())
df[['text','cleaned_text']].head()
```

### Tokenizing

tokenizing is performed, which is the process of selecting, cutting, and truncating words in the document that will be used as terms based on spaces.

```python
df['cleaned_text'] = df['cleaned_text'].apply(lambda x: x.split())
df[['text','cleaned_text']].head()
```

### Stopword Removal 

Stopword removal involves filtering out common words (like "the", "is", "and") that do not carry significant meaning in the context of sentiment analysis. Removing these words reduces noise and improves the quality of the text for analysis.

```python
from nltk.corpus import stopwords
import nltk

# Define stopwords (you can customize or extend this list)
stop_words = set(stopwords.words('english'))

# Function to remove stopwords
def remove_stopwords(tokens):
    return [word for word in tokens if word not in stop_words]

# Apply stopword removal
df['cleaned_text'] = df['cleaned_text'].apply(remove_stopwords)
df[['text','cleaned_text']].head()
```

### Stemming

Stemming reduces words to their root form by removing prefixes and suffixes. This step is useful for normalizing variations of a word (e.g., "running" → "run") to improve consistency in text data.

```python
from nltk.stem import PorterStemmer

stemmer = PorterStemmer()

# Function to apply stemming
def stem_tokens(tokens):
    return [stemmer.stem(word) for word in tokens]

# Apply stemming to the filtered_tokens column
df['cleaned_text'] = df['cleaned_text'].apply(stem_tokens)
df[['text','cleaned_text']].head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import accuracy_score, f1_score, confusion_matrix

def evaluate_model(y_true, y_pred, class_names=None):
    """
    Evaluates the model by calculating accuracy, F1 score, and plotting the confusion matrix.

    Parameters:
        y_true (array-like): True labels.
        y_pred (array-like): Predicted labels.
        class_names (list, optional): List of class names for the confusion matrix. Defaults to None.

    Returns:
        None
    """
    # Calculate metrics
    accuracy = accuracy_score(y_true, y_pred)
    f1 = f1_score(y_true, y_pred, average='weighted')
    
    # Print metrics
    print(f"Accuracy: {accuracy:.2f}")
    print(f"F1 Score: {f1:.2f}")
    
    # Compute confusion matrix
    cm = confusion_matrix(y_true, y_pred)

    # Plot confusion matrix
    plt.figure(figsize=(8, 6))
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=class_names, yticklabels=class_names)
    plt.xlabel('Predicted Labels')
    plt.ylabel('True Labels')
    plt.title('Confusion Matrix')
    plt.show()
```

```python
import xgboost as xgb
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, f1_score, confusion_matrix
from sklearn.preprocessing import LabelEncoder
import matplotlib.pyplot as plt
import seaborn as sns

# Inspect and preprocess the target variable
print(df['target'].unique())  # Check unique classes

# Remap classes (if binary classification)
df['target'] = df['target'].map({0: 0, 4: 1})  # Adjust mapping as needed
print(df['target'].unique())  # Verify remapped classes

# Alternatively, encode classes automatically (for multi-class classification)
# le = LabelEncoder()
# df['target'] = le.fit_transform(df['target'])

texts = [str(text) for text in df['cleaned_text']]  # Ensure all elements are strings
labels = df['target']  # Target variable

# Split data
X_train, X_test, y_train, y_test = train_test_split(texts, labels, test_size=0.2, random_state=42)

# TF-IDF Vectorization
vectorizer = TfidfVectorizer()
X_train_tfidf = vectorizer.fit_transform(X_train)
X_test_tfidf = vectorizer.transform(X_test)

# Train XGBoost model
xgb_model = xgb.XGBClassifier(use_label_encoder=False, eval_metric='logloss')
xgb_model.fit(X_train_tfidf, y_train)

# Predictions
y_pred = xgb_model.predict(X_test_tfidf)
evaluate_model(y_test, y_pred, class_names=["Negative", "Positive"])
```
